The disease has two distinct clades: clade I (with subclades Ia and Ib) and clade II (with subclades IIa and IIb), with clade II being the most recent mpox that has caused a global outbreak in 2024 (CDC, 2024). The focus of this project is to examine how Machine Learning models can be utilized to accurately diagnose specific disease through classification and detection techniques while considering racial bias. The ongoing global outbreak of clade II has caused more than 100 00 cases in 122 countries, including 115 countries where Mpox was not reported previously (CDC, 2024).

The fatality rate of Mpox can reach up to 11% depending on the strain and the health condition of the affected individual, with the Clade I having a mortality rate of around 3.6% (Mpox Virus: Clade I and Clade II, n.d). There is no specific treatment for Mpox, antiviral medications like tecovirimat (TPOXX) can help mitigate symptoms, but these treatments are not accessible in underprivileged areas (CDC, 2024). With the ongoing violence in the Democratic Republic of Congo, efforts to control Mpox have been severely hindered by ongoing violence, which results in higher transmission rates and delayed responses.

The traditional diagnosis method is PCR testing, and Serological testing, where PCR is the primary method that uses samples from skin lesions and is the most preferred because of its high sensitivity and specificity, while Serological tests detect antibodies, but this test is less reliable due to its cross-reactivity with other orthopoxviruses (Khehra, Padda and Swift, 2023). Both these methods have their limitations, and these include the need for specialized equipment and trained personnel and the potential for false negatives due to viral mutations. Enhancing diagnostic accuracy and accessibility, especially in resource-limited environments, requires the development of a reliable computer-based framework for the detection of Mpox disease. The goal is to provide an accessible, reliable tool that aids healthcare professionals in conflict zones and underprivileged areas

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import os
import warnings
from sklearn.preprocessing import LabelEncoder
warnings.filterwarnings('ignore')

# Configure GPU
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"Using {len(gpus)} GPU(s) with memory growth enabled")
    except RuntimeError as e:
        print(f"GPU configuration error: {e}")
else:
    print("No GPU found, using CPU")

# Label mapping based on your dataset
LABEL_MAPPING = {
    'Chickenpox': 0,
    'Cowpox': 1, 
    'HFMD': 2,
    'Healthy': 3,
    'Measles': 4,
    'Monkeypox': 5
}

REVERSE_LABEL_MAPPING = {v: k for k, v in LABEL_MAPPING.items()}

# Custom callbacks for compatibility
class CustomEarlyStopping:
    def __init__(self, monitor='val_loss', patience=3, restore_best_weights=True):
        self.monitor = monitor
        self.patience = patience
        self.restore_best_weights = restore_best_weights
        self.best_weights = None
        self.best_value = float('inf')
        self.wait = 0
        
    def on_epoch_end(self, model, history):
        if not history.history.get('val_loss'):
            return False
            
        current_value = history.history['val_loss'][-1]
        if current_value < self.best_value:
            self.best_value = current_value
            self.best_weights = model.get_weights()
            self.wait = 0
        else:
            self.wait += 1
            
        if self.wait >= self.patience:
            if self.restore_best_weights and self.best_weights is not None:
                model.set_weights(self.best_weights)
            return True
        return False

class CustomReduceLROnPlateau:
    def __init__(self, monitor='val_loss', factor=0.5, patience=2, min_lr=1e-7):
        self.monitor = monitor
        self.factor = factor
        self.patience = patience
        self.min_lr = min_lr
        self.best_value = float('inf')
        self.wait = 0
        
    def on_epoch_end(self, model, history):
        if not history.history.get('val_loss'):
            return
            
        current_value = history.history['val_loss'][-1]
        if current_value < self.best_value:
            self.best_value = current_value
            self.wait = 0
        else:
            self.wait += 1
            
        if self.wait >= self.patience:
            current_lr = tf.keras.backend.get_value(model.optimizer.lr)
            new_lr = max(current_lr * self.factor, self.min_lr)
            tf.keras.backend.set_value(model.optimizer.lr, new_lr)
            print(f"Reducing learning rate to {new_lr:.7f}")
            self.wait = 0

class FederatedLearningClient:
    def __init__(self, client_id, data, image_size=(224, 224), batch_size=32):
        self.client_id = client_id
        self.data = data
        self.image_size = image_size
        self.batch_size = batch_size
        self.model = None
        self.history = []
        self.class_weights = None
        
    def create_model(self, input_shape=(224, 224, 3), num_classes=6):
        """Create DenseNet201 model for medical image classification"""
        # Create base model
        base_model = DenseNet201(
            weights='imagenet',
            include_top=False,
            input_shape=input_shape
        )
        
        # Freeze base model initially
        base_model.trainable = False
        
        # Add custom layers optimized for medical images
        x = base_model.output
        x = GlobalAveragePooling2D()(x)
        x = Dense(1024, activation='relu')(x)
        x = BatchNormalization()(x)
        x = Dropout(0.5)(x)
        x = Dense(512, activation='relu')(x)
        x = BatchNormalization()(x)
        x = Dropout(0.3)(x)
        predictions = Dense(num_classes, activation='softmax')(x)
        
        model = Model(inputs=base_model.input, outputs=predictions)
        
        # Compile with lower learning rate for fine-tuning
        model.compile(
            optimizer=Adam(learning_rate=0.0001),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        
        self.model = model
        return model
    
    def calculate_class_weights(self, y):
        """Calculate class weights for imbalanced medical data"""
        classes = np.unique(y)
        weights = compute_class_weight(
            class_weight='balanced',
            classes=classes,
            y=y
        )
        self.class_weights = dict(zip(classes, weights))
        print(f"Client {self.client_id} class weights: {self.class_weights}")
        return self.class_weights
    
    def prepare_data(self):
        """Prepare client data for training with medical image preprocessing"""
        # Filter data for this client
        client_data = self.data[self.data['Fold'].str.contains(f'fold{self.client_id}_')]
        
        # Use only training data for client training
        train_data = client_data[client_data['Split'] == 'Train']
        
        if len(train_data) == 0:
            print(f"Warning: No training data for client {self.client_id}")
            return None, None
        
        # Create validation split
        train_data, val_data = train_test_split(
            train_data, 
            test_size=0.15,
            random_state=42, 
            stratify=train_data['Updated_Diagnosis_Label']
        )
        
        # Calculate class weights for imbalanced medical data
        self.calculate_class_weights(train_data['Updated_Diagnosis_Label'])
        
        print(f"Client {self.client_id}: {len(train_data)} training, {len(val_data)} validation samples")
        
        # Medical image specific data augmentation
        train_datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=15,
            width_shift_range=0.1,
            height_shift_range=0.1,
            horizontal_flip=True,
            zoom_range=0.1,
            brightness_range=[0.9, 1.1],
            fill_mode='nearest'
        )
        
        val_datagen = ImageDataGenerator(rescale=1./255)
        
        try:
            train_generator = train_datagen.flow_from_dataframe(
                train_data,
                x_col='Image_Path',
                y_col='Updated_Diagnosis_Label',
                target_size=self.image_size,
                batch_size=self.batch_size,
                class_mode='raw',
                shuffle=True
            )
            
            val_generator = val_datagen.flow_from_dataframe(
            val_data,
            x_col='Image_Path',
            y_col='Updated_Diagnosis_Label',
            target_size=self.image_size,
            batch_size=self.batch_size,
            class_mode='raw',
            shuffle=False  # Important: don't shuffle for evaluation
        )
            
            return train_generator, val_generator
            
        except Exception as e:
            print(f"Error creating data generators for client {self.client_id}: {e}")
            return None, None
    
    def train_local_model(self, global_weights=None, epochs=2):
        """Train local model and return updated weights"""
        if global_weights is not None:
            self.model.set_weights(global_weights)
        
        train_gen, val_gen = self.prepare_data()
        
        if train_gen is None:
            print(f"Skipping client {self.client_id} - no data available")
            return None, 0
        
        # Custom callbacks for medical image training
        early_stopping = CustomEarlyStopping(patience=3)
        reduce_lr = CustomReduceLROnPlateau(patience=2)
        
        # Train the model
        try:
            round_history = {'loss': [], 'accuracy': [], 'val_loss': [], 'val_accuracy': []}
            
            for epoch in range(epochs):
                print(f"  Client {self.client_id} - Epoch {epoch + 1}/{epochs}", end=' ')
                
                # Train for one epoch
                history = self.model.fit(
                    train_gen,
                    epochs=1,
                    validation_data=val_gen,
                    verbose=0,
                    class_weight=self.class_weights,
                    steps_per_epoch=min(30, len(train_gen)),  # Conservative steps
                    validation_steps=min(15, len(val_gen)) if val_gen else None
                )
                
                # Store history
                for key in round_history.keys():
                    if key in history.history:
                        round_history[key].extend(history.history[key])
                
                # Check callbacks
                stop_training = early_stopping.on_epoch_end(self.model, history)
                reduce_lr.on_epoch_end(self.model, history)
                
                # Print progress
                current_loss = history.history['loss'][0]
                current_acc = history.history['accuracy'][0]
                print(f"- Loss: {current_loss:.4f}, Acc: {current_acc:.4f}")
                
                if stop_training:
                    print(f"  Early stopping triggered for client {self.client_id}")
                    break
            
            self.history.append(round_history)
            
            # Return updated weights and sample size
            return self.model.get_weights(), len(train_gen.filenames)
            
        except Exception as e:
            print(f"Error training client {self.client_id}: {e}")
            return None, 0


    def evaluate_local_model(self, detailed=False):
        """Evaluate local model performance with optional detailed metrics"""
        _, val_gen = self.prepare_data()
        
        if val_gen is None:
            return None, None, None
            
        # Basic evaluation
        results = self.model.evaluate(val_gen, verbose=0)
        if isinstance(results, list):
            loss, accuracy = results[0], results[1]
        else:
            loss, accuracy = results, 0.0
        
        if detailed:
            # Get true labels and predictions for detailed metrics
            val_gen.reset()  # Reset generator to ensure we start from the beginning
            y_true = []
            y_pred = []
            
            # Iterate through all batches
            for i in range(len(val_gen)):
                batch_x, batch_y = val_gen[i]
                batch_pred = self.model.predict(batch_x, verbose=0)
                y_true.extend(batch_y)
                y_pred.extend(np.argmax(batch_pred, axis=1))
                
                # Break if we've processed all samples
                if (i + 1) * self.batch_size >= len(val_gen.filenames):
                    break
            
            y_true = np.array(y_true)
            y_pred = np.array(y_pred)
            
            # Classification report
            report = classification_report(
                y_true, 
                y_pred, 
                target_names=[REVERSE_LABEL_MAPPING[i] for i in range(6)], 
                output_dict=True
            )
            
            return loss, accuracy, report
        else:
            return loss, accuracy, None
    

class FederatedLearningServer:
    def __init__(self, num_classes=6):
        self.num_classes = num_classes
        self.global_model = self.create_global_model()
        self.global_history = {
            'round': [], 
            'val_loss': [], 
            'val_accuracy': [],
            'client_contributions': []
        }
        
    def create_global_model(self):
        """Create global DenseNet201 model for medical image classification"""
        base_model = DenseNet201(
            weights='imagenet',
            include_top=False,
            input_shape=(224, 224, 3)
        )
        base_model.trainable = False
        
        x = base_model.output
        x = GlobalAveragePooling2D()(x)
        x = Dense(1024, activation='relu')(x)
        x = BatchNormalization()(x)
        x = Dropout(0.5)(x)
        x = Dense(512, activation='relu')(x)
        x = BatchNormalization()(x)
        x = Dropout(0.3)(x)
        predictions = Dense(self.num_classes, activation='softmax')(x)
        
        model = Model(inputs=base_model.input, outputs=predictions)
        
        model.compile(
            optimizer=Adam(learning_rate=0.0001),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        
        return model
    
    def federated_averaging(self, client_weights, sample_sizes):
        """Perform federated averaging with client contribution tracking"""
        print("Performing federated averaging...")
        
        # Filter out None values (failed clients)
        valid_weights = []
        valid_samples = []
        
        for weights, samples in zip(client_weights, sample_sizes):
            if weights is not None and samples > 0:
                valid_weights.append(weights)
                valid_samples.append(samples)
        
        if not valid_weights:
            print("No valid client weights for averaging!")
            return self.global_model.get_weights(), []
        
        print(f"Averaging weights from {len(valid_weights)} clients")
        
        # Calculate total samples and client contributions
        total_samples = sum(valid_samples)
        client_contributions = [samples / total_samples for samples in valid_samples]
        
        # Initialize averaged weights with zeros
        averaged_weights = []
        for weights_list in valid_weights[0]:
            averaged_weights.append(np.zeros_like(weights_list))
        
        # Weighted averaging
        for client_weight, num_samples in zip(valid_weights, valid_samples):
            weight_factor = num_samples / total_samples
            
            for i in range(len(averaged_weights)):
                averaged_weights[i] += client_weight[i] * weight_factor
        
        return averaged_weights, client_contributions
    
    def train_federated_model(self, clients, communication_rounds=10, local_epochs=2):
        """Train model using federated learning with medical image focus"""
        print("Starting Federated Learning for Medical Image Classification...")
        print(f"Disease classes: {list(LABEL_MAPPING.keys())}")
        print(f"Number of clients: {len(clients)}")
        print(f"Communication rounds: {communication_rounds}")
        print(f"Local epochs per round: {local_epochs}")
        
        # Initial evaluation
        print("\n--- Initial Evaluation ---")
        initial_results = self.evaluate_global_model(clients[0])
        if initial_results:
            loss, accuracy, _ = initial_results
            print(f"Initial - Val Loss: {loss:.4f}, Val Accuracy: {accuracy:.4f}")
        
        for round_num in range(communication_rounds):
            print(f"\n--- Communication Round {round_num + 1}/{communication_rounds} ---")
            
            client_weights = []
            sample_sizes = []
            
            # Get global weights
            global_weights = self.global_model.get_weights()
            
            # Train each client
            successful_clients = 0
            for i, client in enumerate(clients):
                print(f"Training Client {client.client_id}...")
                weights, num_samples = client.train_local_model(
                    global_weights=global_weights,
                    epochs=local_epochs
                )
                if weights is not None:
                    client_weights.append(weights)
                    sample_sizes.append(num_samples)
                    successful_clients += 1
                    print(f"Client {client.client_id} completed with {num_samples} samples")
                else:
                    print(f"Client {client.client_id} failed training")
            
            # Perform federated averaging
            if client_weights:
                new_global_weights, contributions = self.federated_averaging(client_weights, sample_sizes)
                
                # Update global model
                self.global_model.set_weights(new_global_weights)
                
                # Evaluate global model
                results = self.evaluate_global_model(clients[0])
                if results:
                    val_loss, val_accuracy, _ = results
                    self.global_history['round'].append(round_num + 1)
                    self.global_history['val_loss'].append(val_loss)
                    self.global_history['val_accuracy'].append(val_accuracy)
                    self.global_history['client_contributions'].append(contributions)
                    
                    print(f"Round {round_num + 1} - Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}")
                    print(f"Successful clients: {successful_clients}/{len(clients)}")
                else:
                    print(f"Round {round_num + 1} - Evaluation skipped")
            else:
                print(f"Round {round_num + 1} - No successful client training")
    
    def evaluate_global_model(self, client, detailed=False):
        """Evaluate global model using client's validation data"""
        try:
            # Temporarily set client model to global weights for evaluation
            original_weights = client.model.get_weights()
            client.model.set_weights(self.global_model.get_weights())
            
            results = client.evaluate_local_model(detailed=detailed)
            
            # Restore original weights
            client.model.set_weights(original_weights)
            
            return results
        except:
            return None

def analyze_medical_dataset(df):
    """Analyze the medical dataset with disease-specific insights"""
    print("=== Medical Dataset Analysis ===")
    print(f"Total samples: {len(df):,}")
    print(f"Number of disease classes: {len(LABEL_MAPPING)}")
    
    print("\nDisease Class Distribution:")
    class_dist = df['Updated_Diagnosis_Label'].value_counts().sort_index()
    for label_idx, count in class_dist.items():
        disease_name = REVERSE_LABEL_MAPPING.get(label_idx, f"Unknown_{label_idx}")
        print(f"  {disease_name} ({label_idx}): {count:>6} samples ({count/len(df)*100:.2f}%)")
    
    print("\nFold Distribution:")
    print(df['Fold'].value_counts().head(10))  # Show top 10 folds
    
    print("\nSplit Distribution:")
    print(df['Split'].value_counts())

def plot_detailed_results(global_history, clients):
    """Plot detailed federated learning results"""
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    # Validation loss and accuracy
    ax1.plot(global_history['round'], global_history['val_loss'], 'b-', marker='o', linewidth=2, label='Validation Loss')
    ax1.set_title('Federated Learning - Validation Loss', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Communication Rounds')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    ax2.plot(global_history['round'], global_history['val_accuracy'], 'r-', marker='o', linewidth=2, label='Validation Accuracy')
    ax2.set_title('Federated Learning - Validation Accuracy', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Communication Rounds')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Client contributions
    if global_history['client_contributions']:
        contributions = np.array(global_history['client_contributions'])
        for i in range(contributions.shape[1]):
            ax3.plot(global_history['round'], contributions[:, i], marker='o', label=f'Client {i+1}')
        ax3.set_title('Client Contributions Over Rounds', fontsize=14, fontweight='bold')
        ax3.set_xlabel('Communication Rounds')
        ax3.set_ylabel('Contribution Ratio')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
    
    # Final class distribution
    if clients and clients[0].class_weights is not None:
        classes = list(clients[0].class_weights.keys())
        weights = list(clients[0].class_weights.values())
        disease_names = [REVERSE_LABEL_MAPPING.get(cls, f"Class {cls}") for cls in classes]
        
        ax4.bar(disease_names, weights, color='skyblue', alpha=0.7)
        ax4.set_title('Class Weights (Inverse Frequency)', fontsize=14, fontweight='bold')
        ax4.set_ylabel('Weight')
        ax4.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.savefig('medical_federated_learning_results.png', dpi=300, bbox_inches='tight')
    plt.show()

def main():
    # Load your dataset
    try:
        # Replace with your actual dataset path
        df = pd.read_csv('/kaggle/input/hsv-color/HSV Color Space Augmentation Dataset/balanced_dataset.csv')
        print("Medical dataset loaded successfully!")


        # === Define new base path ===
        new_base_path = "/kaggle/input/hsv-color/HSV Color Space Augmentation Dataset/balanced_images"
        
        # === Update the Image_Path column ===
        df["Image_Path"] = df["Image_Id"].apply(lambda x: os.path.join(new_base_path, f"{x}.jpg"))
        
        # === Encode the textual Diagnosis column ===
        label_encoder = LabelEncoder()
        df['Updated_Diagnosis_Label'] = label_encoder.fit_transform(df['Diagnosis'])
        
        # Mapping dictionary
        label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
        
        print("✅ Image paths updated and Diagnosis labels encoded successfully")
        print("Label Mapping:")
        print(label_mapping)
        
        # OPTIONAL: Save a new CSV for debugging
        df.to_csv("updated_file.csv", index=False)
        print("CSV saved as updated_file.csv")

        print("Medical dataset loaded successfully!")
    except Exception as e:
        print(f"Error loading dataset: {e}")
        print("Please update the dataset path in the code.")
        return
    
    # Analyze medical dataset
    analyze_medical_dataset(df)
    
    # Identify available clients (folds)
    available_folds = [fold for fold in df['Fold'].unique() if 'fold' in str(fold)]
    client_ids = list(set([fold.split('_')[0].replace('fold', '') for fold in available_folds]))
    client_ids = sorted([cid for cid in client_ids if cid.isdigit()])
    
    print(f"\nAvailable clients: {client_ids}")
    
    if not client_ids:
        print("No valid clients found! Using default clients 1, 2, 3")
        client_ids = ['1', '2', '3']
    
    # Use available clients
    selected_clients = client_ids[:min(3, len(client_ids))]  # Reduced for stability
    print(f"Selected clients for federated learning: {selected_clients}")
    
    # Create clients
    clients = []
    num_classes = len(LABEL_MAPPING)
    
    for client_id in selected_clients:
        client = FederatedLearningClient(
            client_id=client_id, 
            data=df, 
            batch_size=16,  # Reduced for stability
            image_size=(224, 224)
        )
        client.create_model(num_classes=num_classes)
        clients.append(client)
        print(f"Created Client {client_id} for medical image classification")
    
    # Create server
    server = FederatedLearningServer(num_classes=num_classes)
    
    print(f"\nStarting federated learning for {num_classes} disease classes")
    print(f"Diseases: {list(LABEL_MAPPING.keys())}")
    
    # Train federated model
    server.train_federated_model(
        clients=clients,
        communication_rounds=5,  # Reduced for testing
        local_epochs=1
    )
    
    # Plot detailed results
    if server.global_history['round']:
        plot_detailed_results(server.global_history, clients)
        
        # Save training history
        history_df = pd.DataFrame(server.global_history)
        history_df.to_csv('medical_federated_learning_history.csv', index=False)
        print("Training history saved as 'medical_federated_learning_history.csv'")
    else:
        print("No training history to plot")
    
    # Save the final global model
    try:
        server.global_model.save('medical_federated_densenet201_model.h5')
        print("Final medical model saved as 'medical_federated_densenet201_model.h5'")
    except Exception as e:
        print(f"Error saving model: {e}")
    
    # Final detailed evaluation
    print("\n=== Final Detailed Evaluation ===")
    for client in clients:
        results = client.evaluate_local_model(detailed=True)
        if results and results[0] is not None:
            loss, accuracy, report = results
            print(f"\nClient {client.client_id} - Loss: {loss:.4f}, Accuracy: {accuracy:.4f}")
            if report:
                print("Classification Report:")
                print(pd.DataFrame(report).transpose().round(4))

# Run the main function
if __name__ == "__main__":
    main()
```
